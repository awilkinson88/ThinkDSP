##### Question 2

```python
def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys

amps1 = numpy.zeros(50)
amps2 = numpy.ones(50)
amps = numpy.concatenate((amps1,amps2))
fs = numpy.linspace(0,20,num=100)
times = numpy.linspace(0,.5,len(fs))
ys = synthesize2(amps, fs, times)

thinkplot.plot(times,ys)
```

High pass filter window

```python
amps3 = numpy.ones(50)
amps4 = numpy.zeros(50)
amps = numpy.concatenate((amps3,amps4))
fs = numpy.linspace(0,20,num=100)
times = numpy.linspace(0,.5,len(fs))
ys = synthesize2(amps, fs, times)

thinkplot.plot(times,ys)
```

```python
#convolve windows with impulse
box_ys = numpy.convolve(wave.ys, boxcar, mode='same')
box_sig = thinkdsp.Wave(box_ys, framerate=wave.framerate)
box_spec = box_sig.make_spectrum()

#Compute complex angles, and plot
box_ang = numpy.angle(box_spec.hs)
thinkplot.plot(box_ang)
```

##### Question Four

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

padded = zero_pad([1,-1],100)
dft_window = numpy.fft.rfft(padded)
thinkplot.plot(dft_window)
```

Now I will apply the better recording's transfer function to the saxophone.

```python
response = goodrec.segment(start=0.26, duration=5.0)
response.normalize()
transfer = response.make_spectrum()
transfer.plot()
```

```python
transfer.plot()
thinkplot.config(xscale='log', yscale='log')
```

```python
sax = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
sax.ys = sax.ys[:len(response)]
sax.normalize()
sax.plot()
```
