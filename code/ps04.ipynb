##### Question One

```python
class ComplexSinusoid(thinkdsp.Sinusoid):
    """Represents a complex exponential signal."""
    
    def evaluate(self, ts):
        """Evaluates the signal at the given times.

        ts: float array of times
        
        returns: float wave array
        """
        print(ts)
        i = complex(0, 1)
        phases = PI2 * self.freq * ts + self.offset
        print(phases)
        ys = self.amp * numpy.exp(i * phases)
        return ys
```

```python
def synthesize1(amps, freqs, ts):
    components = [thinkdsp.ComplexSinusoid(freq, amp)
                  for amp, freq in zip(amps, freqs)]
    signal = thinkdsp.SumSignal(*components)
    ys = signal.evaluate(ts)
    return ys

def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys
```

```python
amps = numpy.array([0.6, 0.25, 0.1, 0.05])
freqs = [100, 200, 300, 400]
framerate = 11025

ts = numpy.linspace(0, 1, framerate)
ys = synthesize1(amps, freqs, ts)

thinkplot.plot(ts[:500], ys[:500].real)
thinkplot.plot(ts[:500], ys[:500].imag)
```

```python
def shift_wave(fs,offset):
    shifts = []
    for f in fs:
        shift = numpy.exp(-offset*i*f)
        shifts.append(shift)
    return shifts
```

```python
#takes list of equally spaced samples
#returns: list of coefficients 

def dft (ys):
    N = len(ys)
    f_k = numpy.zeros((N), dtype=numpy.complex128)
    k = 0
    while k<N:
        n = 0
        while n < N:
            f_k[k] += ys[n]*numpy.exp(-2.0*math.pi*1j*n*k/N)
            n += 1
        k += 1
    return f_k
```

```python
def analyze2(ys, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    amps = M.conj().transpose().dot(ys) / N
    return amps

N = 4
amps = numpy.array([0.6, 0.25, 0.1, 0.05])
freqs = numpy.arange(N)
ts = numpy.arange(N) / N
ys = synthesize2(amps, freqs, ts)

amps2 = analyze2(ys, freqs, ts)
print(amps2)
```

```python
print(dft(ys))
```

The Fourier transform implemented above is consistent with Allen Downey's analyze and fft functions. Essentially, it sums up the complex exponentials and compares them to the signal to find the amplitude and strength of each frequency component.

```python
import cmath
import math

def omega(p,q):
    return cmath.exp((2.0*cmath.pi * 1j *q) / p)

def fft(ys):
    n = len(ys)
    if n == 1:
        return ys
    else:
        Feven = fft([ys[i] for i in xrange (0,n,2)])
        Fodd = fft([ys[i] for i in xrange(1,n,2)])
        
        combined = [0] * n
        for m in xrange(int(n/2)):
            combined[m] = Feven[m] + omega(n,-m) * Fodd[m]
            combined [m + int(n/2)] = Feven[m] - omega(n, -m) * Fodd[m]
            
        return combined
    
print (fft(ys))
```

The FFT takes advantage of the fact that the signal is periodic, and splits a signal into smaller pieces. It then computes the Fourier transform of each individual piece and sums them back up.

```python
def serial_corr(wave, lag=1):
    n = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:n-lag]
    corr = numpy.corrcoef(y1, y2)[0, 1]
    return corr

def autocorr(wave):
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

ns = 2 ** numpy.arange(4, 12)
sin = thinkdsp.SinSignal(freq=100,amp=1)
wav = sin.make_wave()

results = []
for N in ns:
    result = %timeit -o autocorr(wav)
    results.append(result)
    
    
```

```python
def zero_pad(array, n):
    """Makes a new array with the same elements and the given length.

    array: numpy array
    n: length of result

    returns: new NumPy array
    """
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = numpy.fft.fft(signal)
    fft_window = numpy.fft.fft(window)
    return numpy.fft.ifft(fft_signal * fft_window)

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)
    corrs = fft_convolve(signal,window)
    corrs = corrs[N//2: 3*N//2]
    return corrs

results2 = []
for N in ns:
    result = %timeit -o fft_autocorr(ys2)
    results2.append(result)
```
