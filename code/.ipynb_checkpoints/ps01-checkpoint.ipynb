```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import numpy
import copy

%matplotlib inline
#thinkdsp.play_wave(filename='wolf.wav', player='aplay')
```

Plays wolf howl recording

```python
wave = thinkdsp.read_wave('wolf.wav')
wave.normalize()
wave.make_audio()
```

```python
start = 2.0
duration = 1.0
segment = wave.segment(start,duration)
segment.apodize()
segment.make_audio()
```

Segment with constant pitch

```python
wave.segment(start, duration=0.01).plot()
thinkplot.config(xlabel ='time(s)', ylim=[-1.05,1.05], legend=False)
```

A few cycles of the waveform, which appears to be a mixture of sinusoids.

```python
spectrum = segment.make_spectrum()
high = 1500
spectrum.plot(high=high)
```

```python
for f, amp in zip(spectrum.fs, spectrum.amps):
    if amp > 350:
        print(f, amp)
```

The harmonic structure of the wolf howl appears to have a fundamental, dominant frequency around 538 Hz, with an additional harmonic around 1003 Hz.

```python
lowfreq = copy.deepcopy(spectrum)
lowfreq.low_pass(cutoff=600, factor=.01)
lowave = lowfreq.make_wave()
lowave.make_audio()
```

```python
highfreq = copy.deepcopy(spectrum)
highfreq.high_pass(cutoff=600, factor=.01)
hiwave = highfreq.make_wave()
hiwave.make_audio()
```

The low pass filter had fewer of the higher harmonics, but did not sound very different. The high pass filter had the same perceived pitch, despite having the fundamental tone removed.

```python
fund = thinkdsp.SinSignal(freq=150, amp=1.5, offset=0)
harm1 = thinkdsp.SinSignal(freq=300, amp=.75, offset=0)
harm2 = thinkdsp.SinSignal(freq=600, amp=.375, offset=0)
harm3 = thinkdsp.SinSignal(freq = 555, amp = .75)
newsig = fund + harm1 + harm2
newav = newsig.make_wave(duration=.2, start = 0, framerate=11025)
period = newsig.period
newavseg = newav.segment(start=0,duration=period*3)
newavseg.plot()
```

```python
newav.make_audio()
```

```python
badsig = fund + harm3
badwav = badsig.make_wave(duration=1, start = 0, framerate=11025)
badwav.make_audio()
```

It sounds dischordant when you add frequency components that are not integer multiples of the fundamental

```python
class SawtoothSignal(thinkdsp.Sinusoid):
    
    def evaluate(self,ts):
        cycles = self.freq * ts  + self.offset / thinkdsp.PI2
        frac, _ = numpy.modf(cycles)
        ys = frac-.5
        return ys
    
sawtooth = SawtoothSignal(freq=440)
toothwav = sawtooth.make_wave(framerate=11025)
toothwav.segment(start=0,duration=sawtooth.period*3).plot()
toothwav.make_audio()
```

```python
toothwav.make_spectrum().plot()
```

How is the Sawtooth spectrum different from that of a square or triangle wave? Triangle signals have harmonics that die off quicker after the fundamental, at a rate of $1/f^2$. Sawtooth and square waves die off at a rate of $1/f$. Triangle and square waves only have odd harmonics. Sawtooth waves have both even and odd harmonics

```python
trisig = thinkdsp.TriangleSignal(1100)
trisig.make_wave(framerate=10000).make_audio()
```

The aliased harmonic is especially obvious when you set the TriangleSignal frequency to 1400 Hz.

```python
sqalias = thinkdsp.SquareSignal(1100)
sqwav = sqalias.make_wave(framerate=10000,duration=2)
spec = sqwav.make_spectrum()
spec.plot()
```

The allsquare wave has more smaller harmonics than the triangle wave from figure 2.5

```python
wav = thinkdsp.SinSignal(freq=100,amp=1/100**2)

for i in range(2,20):
    harmonic = thinkdsp.SinSignal(freq=i*100, amp=1/((i*100)**2))
    wav = wav + harmonic
    
wav10 = wav.make_wave()
wav10.make_spectrum().plot()
wav10.make_audio()
```

```python
wav10seg = wav10.segment(duration=wav.period*3)
wav10seg.plot()
```
