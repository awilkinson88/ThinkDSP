```python
class SawtoothChirp(thinkdsp.Chirp):
    
    def _evaluate(self, ts, freqs):
        dts = numpy.diff(ts)
        dps = thinkdsp.PI2 * freqs * dts
        phases = numpy.cumsum(dps)
        frac, _ = numpy.modf(phases)
        ys = self.amp * (frac-.5)
        return ys
```

```python
signal=SawtoothChirp(start=20,end=240)
wave1= signal.make_wave(duration=2)
wave1.segment(start=0, duration=0.1).plot()
wave1.make_audio()
```

The signal is more stable at lower frequencies, but as the frequency gets higher there is increasingly more aliasing. This is because the framerate is not able to sample the wave enough to accurately recreate it without extra signals.

```python
def glissando():
    signal1 = thinkdsp.Chirp(start=262,end=349)
    signal2 = thinkdsp.Chirp(start=349,end=262)
    wave1 = signal1.make_wave(duration=1)
    wave2 = signal2.make_wave(duration=1)
    gliss = wave1.__or__(wave2)
    return gliss.make_audio()

glissando()
    
```

Q5) Above is the power spectrum of the noise from a rainstorm. It appears to have mostly brownian noise behavior due to the sharp exponential drop as the frequency increases. However, there is an interesting spike around 15000 or 20000 Hz which is unexpected for Brownian noise. This potentially due to additional noise picked up in the recording equipment but not necessarily correlated with the rain noise. 

```python
n = 0
specs = []
for i in range(20):
    seg = noise.segment(duration=.2,start=n)
    spec = seg.make_spectrum()
    spec.amps**2/.2
    specs.append(spec)
    n+=.2

newamps = []
for i in range(len(specs)-1):
    newamps = specs[i].amps + specs[i+1].amps
    
bartlett = thinkdsp.IntegratedSpectrum(cs = newamps, fs = specs[0].fs)
bartlett.plot_power()
thinkplot.config(xlabel ='frequency(Hz)', ylabel = 'power density')
thinkplot.show(xscale='log',yscale='log')
```
